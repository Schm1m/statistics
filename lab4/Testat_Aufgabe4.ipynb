## e) Diagramm Normalverteilung mit Dichte 
Sofern bei einer Messreihe eine Normalverteilung vorliegt, erstellen Sie ein weiteres Diagramm, das sowohl das Histogramm als auch die (an das Histogramm angepasste) Dichtekurve der passenden Normalverteilung zeigt. <br>
Welche Normalverteilungsdichte haben Sie gezeichnet, d. h., welche Werte haben Sie für die Parameter $\mu$ bzw. $\sigma$ eingesetzt?

```python
# copied the creation of the histogram for df2
# then created a normal distribution with the std. deviation and mean
# plotted both in one graph so they are comparable

fig, ax = plt.subplots()

n, bins, patches = plt.hist(
    df2, bins=10, range=(df2.min()[0].round(), df2.max()[0].round())
)
ax.set_xticks(bins)
for rect, label in zip(patches, n):
    height = rect.get_height()
    ax.text(
        rect.get_x() + rect.get_width() / 2,
        height + 0.1,
        int(label),
        ha="center",
        va="bottom",
    )

mu = df2.mean()[0]
std = df2.std()[0]
xmin, xmax = ax.get_xlim()
x= np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mu, std)

ax.plot(x, p*100, 'r') # the factor 100 is choosen, so the graphs are roughly the same size ( in y direction )

ax.set_xlabel("Klassen")
ax.set_ylabel("Anteil der Messwerte in %")
ax.set_title("Histogramm + Dichtefunktion von df2")

plt.show()
```

## g) Verteilungsfunktion
Erstellen Sie für die Messreihe(n), die normalverteilt ist/sind, ein zusätzliches Diagramm, das diejenige Verteilungsfunktion zeigt, die zu der in e) gezeichneten Dichte gehört. 

```python
# the same as creating the graph for the normal distribution, but with a different function

fig, ax = plt.subplots()
c = norm.cdf(x, mu, std)
ax.grid(axis='y')
ax.plot(x, c, "k")
ax.set_xlabel("Werte")
ax.set_ylabel("Wahrscheinlichkeit")
ax.set_title("Verteilungsfunktion von df2")
plt.show()
```
